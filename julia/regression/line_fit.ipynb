```julia
using LinearAlgebra
using VMLS
using Plots

dot(A, B) = A'*B
function line_fit(X, Y)
    XHAT = X .- avg(X)
    m = round(dot(Y, XHAT)/dot(XHAT, XHAT), digits=3)
    b = round(avg(Y) - avg(X)* dot(Y, XHAT)/dot(XHAT, XHAT), digits=3)
    return m, b 
end
```

```julia
a = [(1, 7), (2, 13), (3, 20), (4, 26), (5, 33), (6, 40), (7, 44), (8, 48), (9, 52), (10, 56), (11, 60), 
    (12, 64), (13, 68),(14, 72),(15, 76),(16, 80),(17, 84),(18, 88),(19, 92),(20, 96),(21, 100),(22, 104)]
X = [v[1] for v in a]
Y = [v[2] for v in a]

m, b = line_fit(X, Y)
```

```julia
xseries = linspace(0,23,200)
yseries = m * xseries .+ b

h1 = scatter(X, Y, xlabel="cat years", ylabel="human years", label="\$  (x_i, y_i) \$")
h2 = plot!(xseries, yseries, label ="\$  $m x + $b \$")
```
